```python
import pandas as pd 
```

```python
#  Load the dataset 
df = pd.read_csv('E:\Data Analyst Projects\Customer-Churn-Project\data\Advanced_retail_analytics.csv')
df.head()
```

```python
# check info about data 
df.info()
```

```python
# checking null values
df.isnull().sum()
```

```python
# Removing Duplicates 
df.drop_duplicates(inplace=True)
```

```python
def churn_label(score):
    if score > 70:
        return "High Risk"
    elif score > 40:
        return "Medium Risk"
    else:
        return "Low Risk"

df['churn_category'] = df['churn_risk_score'].apply(churn_label)

```

```python
df.to_csv('E:\Data Analyst Projects\Customer-Churn-Project\data\Advanced_retail_analytics.csv', index=False)
```
